## 2 `Path`

The build-in Python module `pathlib` provides a convenient way to handle file paths

```python
from pathlib import Path

folder_path = Path('path/to/my/data')
file_path = Path('image_0.png')
full_path = folder_path / file_path
```

Additionally, `Path` can be used check if files exist
```python
full_path = Path('path/to/my/data/image_0.png')
full_path.is_file()  # returns True or False
```
create folders
```python
Path('existing_folder/new_folder').mkdir()
```
or extract the filename and filestem out of a filepath
```python
full_path = Path('path/to/my/data/image_0.png')
full_path.name  # returns image_0.png
full_path.stem  # returns image_0
```
